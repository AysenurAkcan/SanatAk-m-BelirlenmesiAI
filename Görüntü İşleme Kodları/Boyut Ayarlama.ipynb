```python
import cv2
import os

# Görsellerin bulunduğu ana klasör yolu belirtildi.
folder_path = r"E:\YazLab Proje-SonHal"
subfolders = ["Cubism", "Expressionism", "Impressionism", "Realism", "Surrealism"]

# Klasörün mevcut olup olmadığını kontrol etme
if os.path.exists(folder_path):
    print("Klasör mevcut.")
else:
    print("Klasör mevcut değil.")

# Görüntülerin boyutu
target_size = (320, 320)

# Alt klasörlerdeki tüm dosyaları listeleme ve boyutlandırma
for subfolder in subfolders:
    subfolder_path = os.path.join(folder_path, subfolder)
    
    if not os.path.exists(subfolder_path):
        print(f"{subfolder} klasörü mevcut değil.")
        continue
    
    for filename in os.listdir(subfolder_path):
        file_path = os.path.join(subfolder_path, filename)
        # Görüntü okuma
        image = cv2.imread(file_path)

        # Görüntünün okunup okunmadığını kontrol etme
        if image is None:
            print(f"{filename} bir görüntü dosyası değil veya okunamıyor.")
            continue

        # Görüntü yeniden boyutlandırıldı.
        resized_image = cv2.resize(image, target_size)

        # Görüntü aynı isimle klasöre geri kaydedildi.
        cv2.imwrite(file_path, resized_image)
        print(f"{filename} başarıyla {target_size} boyutuna getirildi.")

print("Tüm görüntüler yeniden boyutlandırıldı.")
```
